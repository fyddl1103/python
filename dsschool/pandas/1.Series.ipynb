```python
s1 = pd.Series({"서울" : 1241512, "부산" : 15426346, "인천" : 3546734, "대전" : 745754})
s1
```

```python
s2 = pd.Series({"서울" : 7547456, "부산" : 643634, "인천" : 124125},
              index = ["부산", "인천", "서울"]) # index를 리스트로 지정하면 순서 정할 수 있음
s2
```

```python
ds = s2 - s1
ds
```

```python
s1.values - s2.values
```

```python
ds.notnull()
```

```python
ds[ds.notnull()]
```

```python
rs = (s1 - s2) / s2 * 100
rs = rs[rs.notnull()]
rs
```

```python
rs["부산"] = 1.6
```

```python
rs
```

```python
del rs["인천"]
```
