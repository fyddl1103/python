### 기술통계
- 데이터의 개수(count)
- 평균(mean, average)
- 분산(variance)
- 표준 편차(standard deviation)
- 최댓값(maximum)
- 최솟값(minimum)
- 중앙값(median)
- 사분위수(quartile)

```python
x = np.array([18,   5,  10,  23,  19,  -8,  10,   0,   0,   5,   2,  15,   8,
              2,   5,   4,  15,  -1,   4,  -7, -24,   7,   9,  -6,  23, -13])
```

```python
len(x)
```

```python
# 평균
x.mean()
```

```python
# 분산
x.var()
```

```python
# 표준편차
x.std()
```

```python
# 중앙값 
np.median(x)
```

```python
# 사분위수
np.percentile(x, 0)
```

```python
np.percentile(x, 25)
```

```python
np.percentile(x, 75)
```

```python
np.sort(x)
```
