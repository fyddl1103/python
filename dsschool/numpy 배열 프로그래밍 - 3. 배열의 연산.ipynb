#### 넘파이에서는 서로 다른 크기를 가진 두 배열의 사칙연산도 가능. > '브로드캐스팅' 
- 크기가 작은 배열을 자동으로 반복 확장하여 크기가 큰 배열에 맞추는 방법

```python
x = np.arange(5)
x
```

```python
x + 1  # [0,1,2,3,4] + [1,1,1,1,1]
```

```python
m = np.vstack([range(7)[i:i + 3] for i in range(5)])
m
```

```python
l = np.arange(5)[:, np.newaxis]
l
```

```python
m + l
```

#### 차원축소 연산
- 최대/최소: min, max, argmin, argmax
- 통계: sum, mean, median, std, var
- 불리언: all, any

```python
a = np.array([[1,2,3,4],[4,2,2,4],[4,2,2,4],[4,2,2,4]])
```

```python
a
```

```python
a.mean()
```

```python
a.mean(axis=0) # 열합계
```

```python
a.mean(axis=1) # 행합계
```
