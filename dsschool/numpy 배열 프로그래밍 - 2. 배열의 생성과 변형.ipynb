### 배열 연결
행,열의 수가 같은 두 개 이상의 배열을 연결(concatenate) 하여 더 큰 배열을 만듦
- hstack ; 행의 수가 같은 두 개 이상의 배열을 옆으로 연결 (열의 수가 늘어남)
- vstack ; 열의 수가 같은 두 개 이상의 배열을 옆으로 연결 (행의 수가 늘어남)
- dstack ; 깊이 방향으로 배열을 합침.
- stack
- r_
- c_
- tile ; 동일한 배열을 반복하여 연결.

```python
a1 = np.ones((2,3))
a1
```

```python
a2 = np.zeros((2,1))
a2
```

```python
np.hstack([a1, a2])
```

```python
b1 = np.ones((3,2))
b1
```

```python
b2 = np.zeros((5,2))
b2
```

```python
np.vstack([b1, b2])
```

```python
c1 = np.ones((3,5))
c1
```

```python
c2 = np.zeros((3,5))
c2
```

```python
np.dstack([c1,c2])
```

```python
a = np.array([[0,1,2], [3,4,5]])
np.tile(a, 2)
```

```python
np.tile(a, (3,2))
```
