#### 난수(random number) > Numpy의 random 서브패키지에서 제공
- rand() : 0부터 1사이의 균일 분포
- randn() : 표준 정규 분포
- randint() : 균일 분포의 정수 난수

```python
np.random.seed(0)
```

```python
np.random.rand(10) # rand() ; 0과 1사이의 난수를 발생시키는 함수
```

```python
np.random.rand(10)
```

```python
x = np.arange(10)
x
```

```python
np.random.shuffle(x)
x
```

```python
np.random.randint(10, 20, size=(3,10))
```

#### 데이터샘플링
- choice()

```python
np.random.choice(5, 3, p=[0.0, 0, 0.0, 0.1, 0.9]) # p ; 각 수의 선택확률(합 = 1)
```

```python
np.random.choice(5, 5, replace=False) # like shuffle()
```

```python
np.random.choice(5, 10)
```

#### 정수 데이터 카운팅
- unique ; unique 함수는 데이터에서 중복된 값을 제거하고 중복되지 않는 값의 리스트를 출력한다
- bincount ; bincount 함수는 0 부터 minlength - 1 까지의 숫자에 대해 각각 카운트를 한다. 데이터가 없을 경우에는 카운트 값이 0이 된다

```python
# help(np.unique)
```

```python
np.unique([11, 11, 13, 15, 2, 5, 2])
```

```python
a = np.array([11, 11, 13, 15, 2, 5, 2])
val, cnt = np.unique(a, return_counts=True)
```

```python
val
```

```python
cnt
```

```python
x = np.repeat(val, cnt)
x
```

```python
np.bincount(x, minlength=20)
```
