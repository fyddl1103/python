```python
sec_dpc.name = 'sec'
```

```python
msft_dpc.shift(1)
```

#### 3.4.3 주간 일간 변동률 히스토그램

```python
import matplotlib.pyplot as plt
```

```python
sec_dpc = (sec['Close'] - sec['Close'].shift(1)) / sec['Close'].shift(1) * 100
```

```python
sec_dpc.iloc[0] = 0
```

```python
sec_dpc
```

```python
sec_dpc = (sec['Close'] - sec['Close'].shift(1)) / sec['Close'].shift(1) * 100
sec_dpc.iloc[0] = 0
plt.hist(sec_dpc, bins = 10)
#plt.legend(loc='best')
plt.grid(True)
plt.ylim(0,3)
plt.show()
```

```python
# 해석 ; 정규분포보다 중앙이 더 뾰족 (급첨분포) ; 매우 작은 범위 안에서 변동이 발생함
#        두꺼운 양 끝 꼬리 (펫테일) ; 저/고가에서 가격변동이 평균부분보다 많음
```

```python
sec_dpc.describe()
```

#### 3.4.4 일간 변동률 누적합 구하기

```python
# 종목별로 전체적인 변동률 보고 싶을땐 ? 일간 변동률의 누적합 > cumsum()
```

```python
sec_dpc_cs = sec_dpc.cumsum()
```

```python
sec_dpc_cs
```

```python
msft_dpc_cs = msft_dpc.cumsum()
```

```python
msft_dpc_cs
```

```python
plt.plot(sec.index, sec_dpc_cs, 'b', label='samsung')
plt.plot(msft.index, msft_dpc_cs, 'r--', label='Microsoft')
plt.ylabel('Change %')
plt.grid(True)
plt.legend(loc='best')
plt.show()
```
